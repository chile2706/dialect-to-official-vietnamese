# Nghe An

```python
import whisper
model = whisper.load_model("small") # automatically load to cuda:0
```

```python
model.device
```

```python
df = pd.read_parquet(path+"nghe_an.parquet")
```

```python
transcription = []
for i in range(df.shape[0]):
  print(i)
  audio_tensor = torch.from_numpy(np.array(df["audio"][i]["array"])).float()
  # audio_tensor = audio_tensor / np.max(np.abs(audio_tensor))
  audio_tensor = audio_tensor.to(model.device)
  result = model.transcribe(audio_tensor, language="vi")
  print(result["text"])
  transcription.append(result["text"])
```

```python
df['whisper_transcript'] = transcription
df.to_csv(path+"whisper_nghe_an_transcription.csv")
```
